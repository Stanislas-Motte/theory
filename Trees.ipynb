## Decision Trees : Classification

#### What is a Decision Tree
A decision tree is a common ML algo used to classify or regress data. It searches through all features and splits the data to gian the most information on it until it can no longer gain information. 

Training this algorithm creates **a set of decisions** one can take based on the parameters to fit a new instance (observation) directly into a class. 



## Important concepts for DTs

#### Information Gain
Measures by how much we are able to reduce entropy, i.e the expectation of surprise in a dataset. 

The amount of information that a feature gives us about the diffrent classes. 

#### Gini Impurity
A measure of purity for a leaf, i.e how good the split was at separating the data in the parent node. 

$$ Gini Impurity = 1 - p_1^2 - p_2^2... -p_n^2 $$

A perfectly pure leaf will have a Gini Impurity of $0$.

#### Entropy

Entropy is the expectation of surprise, denoted mathematically as: 

$$\sum x P(X=x) $$
where: 
- $x$ is the value of surprise for a given outcome
- $P(X=x)$ is the probability of oberving that specific value for Surprise

which can be written 

$$Entropy = \sum log(\frac{1}{p(x)}) p(x) = - \sum p(x)log(p(x))$$


[Watch everyone's favorite Josh Starmer for an intuitive explanation](https://www.youtube.com/watch?v=YtebGVx-Fxw)



```python
import matplotlib.pyplot as plt
import numpy as np

def classification_error(p):
   return 1 - np.max([p, 1 - p])
   
def gini(p):
   return (p)*(1 - (p)) + (1 - p)*(1 - (1-p))

def entropy(p):
   return - p*np.log2(p) - (1 - p)*np.log2((1 - p))

x = np.arange(0.0, 1.0, 0.01)
ent = [entropy(p) if p != 0 else None for p in x]
scaled_ent = [e*0.5 if e else None for e in ent]
c_err = [classification_error(i) for i in x]

fig = plt.figure()
ax = plt.subplot(111)

for j, lab, ls, c, in zip(
      [ent, scaled_ent, gini(x), c_err],
      ['Entropy', 'Entropy (scaled)', 'Gini Impurity', 'Misclassification Error'],
      ['-', '-', '--', '-.'],
      ['lightgray', 'red', 'green', 'blue']):
   line = ax.plot(x, j, label=lab, linestyle=ls, lw=1, color=c)

ax.legend(loc='upper left', bbox_to_anchor=(0.01, 0.85),
         ncol=1, fancybox=True, shadow=False)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')

plt.ylim([0, 1.1])
plt.xlabel('p(j=1)')
plt.ylabel('Impurity Index')
plt.show()
```

### What kind of model is a DT? 
1. Supervised learning algorithm: labeled data
2. Classification/Regression: can predict classes or values
3. Non parametric: takes no parameters


### What are some advantages/disadvantages of a DT?
**Advantages**
1. Outliers have no effect: each instance counts equally
2. Can classify and regress
3. Does not require any parameter tuning
4. Can be visualized well
**Disadvantages**
1. Low Bias, High Variance (bad at generalizing)
